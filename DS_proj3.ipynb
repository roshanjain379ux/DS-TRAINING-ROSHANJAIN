```python
# IPL MATCH DATA EXPLORER PROJECT

# STEP 1: IMPORT LIBRARIES
import pandas as pd
import numpy as np

# STEP 2: LOAD DATASET
file_path = "/root/.cache/kagglehub/datasets/manasgarg/ipl/versions/5/matches.csv"

# Load dataset
df = pd.read_csv(file_path)

print("===================================")
print("IPL DATASET LOADED SUCCESSFULLY")
print("===================================")

# STEP 3: BASIC DATA EXPLORATION
print("\n===================================")
print("FIRST 5 ROWS")
print("===================================")

print(df.head())

print("\n===================================")
print("DATASET SHAPE")
print("===================================")

print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")

print("\n===================================")
print("COLUMN NAMES")
print("===================================")

print(df.columns)

print("\n===================================")
print("DATASET INFORMATION")
print("===================================")

print(df.info())

print("\n===================================")
print("STATISTICAL SUMMARY")
print("===================================")

print(df.describe(include='all'))

# STEP 4: MATCHES PLAYED EACH SEASON
print("\n===================================")
print("MATCHES PLAYED EACH SEASON")
print("===================================")

matches_per_season = (
    df['season']
    .value_counts()
    .sort_index()
)

print(matches_per_season)

# STEP 5: TEAM WITH MOST WINS
print("\n===================================")
print("TEAM WITH MOST WINS")
print("===================================")

team_wins = df['winner'].value_counts()

print(team_wins)

most_successful_team = team_wins.idxmax()
most_wins = team_wins.max()

print(f"\nMost Successful Team : {most_successful_team}")
print(f"Total Wins           : {most_wins}")

# STEP 6: LIST ALL AVAILABLE VENUES
print("\n===================================")
print("AVAILABLE VENUES")
print("===================================")

venues = df['venue'].unique()

for venue in venues:
    print(venue)

# STEP 7: MATCHES AT A SPECIFIC VENUE
venue_name = "Wankhede Stadium"

venue_matches = df[df['venue'] == venue_name]

print("\n===================================")
print(f"MATCHES PLAYED AT {venue_name.upper()}")
print("===================================")

print(
    venue_matches[
        [
            'season',
            'team1',
            'team2',
            'winner',
            'venue'
        ]
    ]
)

print(f"\nTotal Matches at {venue_name}: {len(venue_matches)}")

# STEP 8: TOP 5 PLAYER OF THE MATCH AWARDS
print("\n===================================")
print("TOP 5 PLAYER OF THE MATCH WINNERS")
print("===================================")

top_players = (
    df['player_of_match']
    .value_counts()
    .head(5)
)

print(top_players)

# STEP 9: CHECK NULL VALUES
print("\n===================================")
print("NULL VALUES IN DATASET")
print("===================================")

null_values = df.isnull().sum()

print(null_values)

# STEP 10: HANDLE MISSING VALUES
print("\n===================================")
print("HANDLING MISSING VALUES")
print("===================================")

# Drop unnecessary column
# umpire3 is fully empty
df = df.drop(columns=['umpire3'])
print("Dropped Column: umpire3")

# Remove rows with important missing values
clean_df = df.dropna(
    subset=[
        'winner',
        'player_of_match',
        'city'
    ]
)

print("\nDataset Cleaned Successfully")

print(f"\nOriginal Shape : {df.shape}")
print(f"Cleaned Shape  : {clean_df.shape}")

# STEP 11: SORT TEAMS BY TOTAL WINS
print("\n===================================")
print("TEAMS SORTED BY TOTAL WINS")
print("===================================")

sorted_team_wins = (
    clean_df['winner']
    .value_counts()
    .sort_values(ascending=False)
)

print(sorted_team_wins)

# STEP 12: FILTER MATCHES
print("\n===================================")
print("MATCHES WON BY MUMBAI INDIANS")
print("===================================")

mi_matches = clean_df[
    clean_df['winner'] == 'Mumbai Indians'
]

print(
    mi_matches[
        [
            'season',
            'team1',
            'team2',
            'winner'
        ]
    ].head(10)
)

# STEP 13: GROUPBY ANALYSIS
print("\n===================================")
print("AVERAGE WIN BY RUNS EACH SEASON")
print("===================================")

average_runs = (
    clean_df
    .groupby('season')['win_by_runs']
    .mean()
)

print(average_runs)

# STEP 14: BONUS CHALLENGE WIN PERCENTAGE ANALYSIS
print("\n===================================")
print("WIN PERCENTAGE ANALYSIS")
print("===================================")

# TEAMS WINNING AFTER BATTING FIRST
batting_first = clean_df[
    clean_df['win_by_runs'] > 0
]

batting_first_wins = (
    batting_first['winner']
    .value_counts()
)

# TEAMS WINNING AFTER FIELDING FIRST

fielding_first = clean_df[
    clean_df['win_by_wickets'] > 0
]

fielding_first_wins = (
    fielding_first['winner']
    .value_counts()
)

# TOTAL TEAM WINS
total_wins = (
    clean_df['winner']
    .value_counts()
)

# CALCULATE PERCENTAGES
batting_first_percentage = (
    batting_first_wins
    /
    total_wins
) * 100

fielding_first_percentage = (
    fielding_first_wins
    /
    total_wins
) * 100

# Replace NaN with 0
batting_first_percentage = (
    batting_first_percentage
    .fillna(0)
    .round(2)
)

fielding_first_percentage = (
    fielding_first_percentage
    .fillna(0)
    .round(2)
)

# DISPLAY RESULTS
print("\n-----------------------------------")
print("WIN % WHEN BATTING FIRST")
print("-----------------------------------")

print(
    batting_first_percentage
    .sort_values(ascending=False)
)

print("\n-----------------------------------")
print("WIN % WHEN FIELDING FIRST")
print("-----------------------------------")

print(
    fielding_first_percentage
    .sort_values(ascending=False)
)

# STEP 15: EXTRA INSIGHTS
print("\n===================================")
print("TOP 10 MOST SUCCESSFUL TEAMS")
print("===================================")

print(sorted_team_wins.head(10))

print("\n===================================")
print("MOST COMMON PLAYER OF THE MATCH")
print("===================================")

best_player = (
    clean_df['player_of_match']
    .value_counts()
    .idxmax()
)

best_player_awards = (
    clean_df['player_of_match']
    .value_counts()
    .max()
)

print(f"Player : {best_player}")
print(f"Awards : {best_player_awards}")

# STEP 16: FINAL SUMMARY
print("\n===================================")
print("PROJECT SUMMARY")
print("===================================")

print(f"Total Matches Analysed : {len(clean_df)}")
print(f"Total Seasons          : {clean_df['season'].nunique()}")
print(f"Total Teams            : {len(clean_df['winner'].unique())}")
print(f"Most Successful Team   : {most_successful_team}")

# PROJECT COMPLETED
print("\n===================================")
print("IPL DATA ANALYSIS COMPLETED")
print("===================================")
```
